 # Exploración inicial del proyecto

```python
import pandas as pd 

file_path = r"C:\Users\apicn\Proyectos\Python_notes\Proyecto analisis denuncias\ConocimientoMediosDenuncia.csv"

data = pd.read_csv(file_path, encoding='latin-1')

# Mostremos una descripción general del DataFrame 

summary = {
    "Head" : data.head(),
    "Info" : data.info(buf = None),
    "Null_sum" : data.isnull().sum(),
    "Describe" : data.describe(include = "all")
}
  
print (summary)
```

```python
data.head(5)
```

```python
# Volvamos la fila 3 como encabezado

data.columns = data.iloc[3]

# Formateemos el indice nuevamente

data = data[4:].reset_index(drop=True)

print(data.head(5))
```

```python
# Eliminar la primera columna por su posición
data = data.drop(data.columns[0], axis=1)

# Mostrar el DataFrame actualizado
print(data.columns)
```

```python
# Elimina espacios en blanco
data["Año"] = data["Año"].str.strip()  
# Reemplazar '.00' al final de los strings con vacío
data["Año"] = data["Año"].str.replace(r'\.00$', '', regex=True)

print(data.head(5))
```

```python
data['Año-Trimestre'] = data['Año'] + ' ' + data['Trimestre']
data = data.drop(["Año","Trimestre"], axis =1)
print(data.columns)
```

# Análisis de datos.
Para poder empezar a verificar que canales de denuncia son mas usados y como este uso ha cambiado a lo largo del tiempo, vamos a pasar a porcentajes los resultados.

```python

denuncia_columns = [
    'Denuncia en línea (Página del SAT, por medio del Buzón de Quejas y Denuncias)',
    'Denuncia telefónica (Marca SAT, 55 627 22 728) ',
    'Denuncia por correo electrónico (denuncias@sat.gob.mx)',
    ' Teléfono Rojo ',
    'Denuncia telefónica Quejas y denuncias SAT (55 88 52 22 22, internacional  844 28 73 803)',
    'De forma presencial al jefe inmediato de la oficina del SAT',
    'SAT Móvil (Aplicación para celular, apartado Quejas y Denuncias)',
    'Sistema Integral de Denuncias Ciudadanas (SIDEC) en línea (https://sidec.funcionpublica.gob.mx)',
    'Denuncia en línea contra servidores públicos federales ante la CNDH (www.cndh.org.mx/Queja_Linea)',
    'No sabía que había medios para denunciar/no los conocía'
]


# Volver numéricos los datos de estas columnas
for col in denuncia_columns:
    data[col] = pd.to_numeric(data[col], errors="coerce").fillna(0)

#Calcular el total por fila
data["Total"] = data[denuncia_columns].sum(axis=1)

#Calculamos los porcentajes por fila
for col in denuncia_columns:
    data[col + "(%)"] = (data[col] / data["Total"]) * 100 

print(data.head(5))
print(data.columns)
```
